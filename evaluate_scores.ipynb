```python
import collections

def create_weights_index(weighting_list):
  words_lists = weighting_list
  # initialize lists
  document_number = 1
  # initialize pos_index
  positional_index = {}
  #intialize position
  position = 1
  #positional_index[term](freq,{doc:[index]})

  for list in words_lists:
    for term in list:
      if term not in positional_index: #if term does not exist
        positional_index[term] = 1/12
      else: #if term exists already
        positional_index[term] +=1/12
  return dict(collections.OrderedDict(sorted(positional_index.items())))

```
