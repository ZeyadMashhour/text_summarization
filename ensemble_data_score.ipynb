```python
refrence_summary_dataset = pd.read_csv(refrence_summary_dataset_path)
#droppping row number 24
refrence_summary_dataset = refrence_summary_dataset.drop(24)
```

```python
system_summaries = pd.read_csv("DUC 5 sentences.csv")
system_summaries = system_summaries.drop(49)
```

```python
combinations = list(system_summaries.columns)
```

```python
def get_combination(combination_name):
    df = pd.DataFrame(system_summaries[combination_name])
    return df
```

```python
avg_scores = []
original_summary = pd.DataFrame(refrence_summary_dataset["Original Summary"])
for combination in tqdm(combinations):
    combination_df = get_combination(combination)#create dataframe
    ensemble_scores = rouge_scores_df(df=original_summary, algorithm_summary_df=combination_df)
    combination_avg = df_avg_by_column(ensemble_scores)
    avg_scores.append(combination_avg)
```

```python
avg_scores_df = pd.DataFrame(avg_scores).T
avg_scores_df.columns = combinations
```

```python
avg_scores_df.to_csv(scores_file_name)
```
