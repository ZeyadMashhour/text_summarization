```python
def get_max_values(df):
    max_values = df.iloc[:, 3:].max(axis=1)
    return max_values
max_values = get_max_values(df)
```

```python
def create_datafrme(df, max_values):
    list_34an_5atr_bassel = []
    for i in range(len(df.index)):
        list_34an_5atr_bassel.append([df.index[i]
            ,max_values[i]
            ,df.columns[list(np.where(df.loc[df.index[i]] == max_values[i]))[0][0]]])
    columns_34an_5atr_hamed = ["metric", "max_score", "algorithm"]
    dataframe_34an_5atr_bassel = pd.DataFrame(list_34an_5atr_bassel, columns=columns_34an_5atr_hamed)
    return dataframe_34an_5atr_bassel
```

```python
files = ['5.csv', '6.csv', '7.csv']
for path in paths:
    for file in files:
        df = pd.read_csv(f"{path}/{file}", index_col= 0)
        max_values = get_max_values(df)
        print(create_datafrme(df, max_values))
```
