```python
def extracted_number(string_with_number):
    if isinstance(string_with_number, str):
        return ''.join(filter(lambda x: x.isdigit() or x == '.', string_with_number))
    return string_with_number
```

```python
#refrence_scores[refrence number[1-4]][summarize percen[5-6-7]]
df_scores_path = f"{refrence_path[refrence_index]}/{refrence_scores[refrence_index][sentence_length]}"
df = pd.read_csv(df_scores_path, index_col= 0)
df.head(12)
```

```python
def extracted_number_from_df(df):
    for column_name in df:
        for i,cell in enumerate(df[column_name]):
            df[column_name][i] = float(extracted_number(cell))
    return df
```

```python
def create_dataframe(df, max_values):
    list_34an_5atr_bassel = []
    for i in range(len(df.index)):
        list_34an_5atr_bassel.append([df.index[i]
            ,max_values[i]
            ,df.columns[list(np.where(df.loc[df.index[i]] == max_values[i]))[0][0]]])
    columns_34an_5atr_hamed = ["metric", "max_score", "algorithm"]
    dataframe_34an_5atr_bassel = pd.DataFrame(list_34an_5atr_bassel, columns=columns_34an_5atr_hamed)
    return dataframe_34an_5atr_bassel
```

```python
create_dataframe(df, max_values)
```
